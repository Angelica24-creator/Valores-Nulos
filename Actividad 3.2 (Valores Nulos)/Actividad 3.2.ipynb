```python
#Importar librerias 

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#Cargar desde un archivo .csv sin indice
df = pd.read_csv("Rio de Janeiro.csv", encoding="latin1")
df
```

```python
df.info()
```

**IDENTIFICAR VALORES NULOS**

```python
#Identificar valores nulos por columnas
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#identificar valores nulos por dataframe

valores_nulos=df.isnull().sum().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe 

data2=df.copy()
```

**Sustituir valores nulos por un string en concreto**

Descripción: Las columnas que tengan entradas de texto ocupare la sustitución por un string concreto, las columnas serán 11: description, neighborhood_overview, host_about, host_response_time, host_is_superhost, neighbourhood, bathrooms_text, calendar_updated, has_availability, last_review y license. Todos los valores nulos los sustituira una leyenda que diga "SIN INFORMACION"

```python
data2["description"]=data2["description"].fillna("SIN INFORMACION")
data2["neighborhood_overview"]=data2["neighborhood_overview"].fillna("SIN INFORMACION")
data2["host_about"]=data2["host_about"].fillna("SIN INFORMACIO")
data2["host_response_time"]=data2["host_response_time"].fillna("SIN INFORMACION")
data2["host_is_superhost"]=data2["host_is_superhost"].fillna("SIN INFORMACION")
data2["neighbourhood"]=data2["neighbourhood"].fillna("SIN INFORMACION")
data2["bathrooms_text"]=data2["bathrooms_text"].fillna("SIN INFORMACION")
data2["calendar_updated"]=data2["calendar_updated"].fillna("SIN INFORMACION")
data2["has_availability"]=data2["has_availability"].fillna("SIN INFORMACION")
data2["last_review"]=data2["last_review"].fillna("SIN INFORMACION")
data2["license"]=data2["license"].fillna("SIN INFORMACION")

```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

**Sustituir valores nulos por un valor numerico en concreto**

Descripción: en las columnas beds, bathrooms y bedrooms ocupare el metodo de sustitución por un numero concreto el cual será 0

```python
data2["beds"]=data2["beds"].fillna(0)
data2["bathrooms"]=data2["bathrooms"].fillna(0)
data2["bedrooms"]=data2["bedrooms"].fillna(0)
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

**Sustituir valores nulos con mediana**

Descripción: Para las columnas host_response_rate y host_acceptance_rate ocupare la mediana para que me de el valor de enmedio de los datos o en dado caso el promedio de los numeros sobrantes.

```python
data2["host_response_rate"]=data2["host_response_rate"].fillna(round(df["host_response_rate"].median(),1))
data2["host_acceptance_rate"]=data2["host_acceptance_rate"].fillna(round(df["host_acceptance_rate"].median(),1))
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

**Sustituir valores nulos con promedio o medio**

Descripción: Ocupe este método en 8 columnas con valores nulos: price, review_scores_accuracy, review_scores_checkin, review_scores_cleanliness, review_scores_communication, review_scores_location, review_scores_value, reviews_per_month y review_scores_rating para que me diera los promedios y con ellos llenas los valores nulos.

```python
data2["price"]=data2["price"].fillna(round(df["price"].mean(),1))
data2["review_scores_accuracy"]=data2["review_scores_accuracy"].fillna(round(df["review_scores_accuracy"].mean(),1))
data2["review_scores_checkin"]=data2["review_scores_checkin"].fillna(round(df["review_scores_checkin"].mean(),1))
data2["review_scores_cleanliness"]=data2["review_scores_cleanliness"].fillna(round(df["review_scores_cleanliness"].mean(),1))
data2["review_scores_communication"]=data2["review_scores_communication"].fillna(round(df["review_scores_communication"].mean(),1))
data2["review_scores_location"]=data2["review_scores_location"].fillna(round(df["review_scores_location"].mean(),1))
data2["review_scores_value"]=data2["review_scores_value"].fillna(round(df["review_scores_value"].mean(),1))
data2["reviews_per_month"]=data2["reviews_per_month"].fillna(round(df["reviews_per_month"].mean(),1))
data2["review_scores_rating"]=data2["review_scores_rating"].fillna(round(df["review_scores_rating"].mean(),1))
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Verificar si aún hay valores nulos por dataframe

valores_nulos=data2.isnull().sum().sum()
valores_nulos
```

```python
#Convertir dataframe a csv
data2.to_csv("Rio de Janeiro_sin_nulos.csv")
```
